## matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

#  1. prepare data 
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10)) #width and height

# 3. plot data 
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple Plot", 
      xlabel="x-axis",
      ylabel="y-axis")

#5Save & Show (save the whole figure)
fig.savefig("images/simple_plot.png")
```

## Two options for Subplot


```python
#subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                           ncols=2,
                                           figsize=(10, 5))
#plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_price.keys(), nut_butter_price.values());
ax4.hist(np.random.randn(1000));
```

```python
fig, ax=plt.subplots(nrows=2,
                    ncols=2,
                    figsize=(10,5))
#Plot to each different index
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
```

```python
#Subplot opetion 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5))
# Plot to each different index
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_price.keys(), nut_butter_price.values());
ax[1,1].hist(np.random.randn(1000));
```

##Which mtd should you use? (pyplot vs matplotlib OO method?)
* when plotting something quickly use pylpot 
* When plotting something more advavanced, use the OO method



```python
heart.head()
```

```python
over_50 = heart[heart["age"] > 50]
over_50
```

```python
len(over_50)
```

```python
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target');
```

```python
# OO method mixed with pyplot method
fix, ax = plt.subplots(figsize=(10,5))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax)
#ax.set_xlim([45, 100]);
```

```python
# OO method from scratch
fig, ax=plt.subplots(figsize=(10,6))

#plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);

#Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol");

#Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

#add horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
#Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols =1,
                              figsize=(10,10))

#Add data to ax0
scatter =ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);
#Customize ax0
ax0.set(title="Heart Disease and Cholesterol levels",
       xlabel= "Age",
       ylabel="Cholesterol");
#add legend
ax0.legend(*scatter.legend_elements(), title="Target");
#add a mean line
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");

#Add date to ax1
scatter =ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"]);
#Customize ax1
ax1.set(title="Heart Disease and Thalach",
       xlabel= "Age",
       ylabel="Thalach");
#add legend
ax1.legend(*scatter.legend_elements(), title="Target");
#add a mean line
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");

#add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
#Set the style
plt.style.use('seaborn-whitegrid')
# OO method from scratch
fig, ax=plt.subplots(figsize=(10,6))

#plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter");   #this means colour map, it changes the colour scheme. The value can be summer, plasma etc

#Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol");

#Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

#add horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
# Customizing the y and x axis limitations

#Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols =1,
                              figsize=(10,10))

#Add data to ax0
scatter =ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter");
#Customize ax0
ax0.set(title="Heart Disease and Cholesterol levels",
       xlabel= "Age",
       ylabel="Cholesterol");

# Change the x axis limit
ax0.set_xlim([50, 80])
#add legend
ax0.legend(*scatter.legend_elements(), title="Target");
#add a mean line
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");

#Add date to ax1
scatter =ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"],
                    cmap="winter");
#Customize ax1
ax1.set(title="Heart Disease and Thalach",
       xlabel= "Age",
       ylabel="Thalach");

#Change x axis limit
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])
#add legend
ax1.legend(*scatter.legend_elements(), title="Target");
#add a mean line
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");

#add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
